```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
ticker = yf.Ticker("TSLA")
```

```python
historical_data = ticker.history(period="1y")
historical_data
```

```python
data=pd.DataFrame(historical_data["Close"])
data.rename(columns={"Close":"price_t"}, inplace=True)
data
```

```python
data["return"]=data["price_t"]/data["price_t"].shift(1)-1
data
```

```python
plt.figure(figsize=(12,8))
plt.plot(data['return'])
plt.xlabel("Date", fontsize=12)
plt.ylabel("Return", fontsize=12)
plt.title("Daily Returns on TSLA")
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(data['price_t'])
plt.xlabel("Date", fontsize=12)
plt.ylabel("Price", fontsize=12)
plt.title("Daily Price of TSLA")
plt.show()
```
