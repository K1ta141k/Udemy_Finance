```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
weekly_data = yf.download("AAPL", start="2000-01-01", end="2017-12-31", interval='1wk')
weekly_data
```

```python
weekly_price = pd.DataFrame(weekly_data['Adj Close'])
weekly_price.rename(columns={"Adj Close":"price_t"}, inplace=True)
weekly_price['returns']=weekly_price['price_t'].pct_change(1)
weekly_price
```

```python
weekly_price['expected returns']=weekly_price['returns'].mean()
weekly_price
```

```python
annualized_return = (1+weekly_price['returns'].mean())**52-1 #52 weeks in year
annualized_return
```
